```python
import osmnx as ox  
tags = {
    'highway': "drive_service",
    'railway': True,
    'cycleway': True,
    'footway': "walk",
    'route': 'bus'
}
slovenia_center = [46.051389, 14.506111]
graphs = []
D = ox.graph.graph_from_point(
    slovenia_center, network_type="drive_service", dist=1000, simplify=False
)
graphs.append(D)

W = ox.graph.graph_from_point(
    slovenia_center, network_type="walk", dist=1000, simplify=False)
graphs.append(W)

for g in graphs:
    # turn off strict mode and see what nodes we'd remove, in yellow
    nc = ["r" if ox.simplification._is_endpoint(g, node, None, None) else "y" for node in g.nodes()]
    fig, ax = ox.plot.plot_graph(g, node_color=nc)
```
